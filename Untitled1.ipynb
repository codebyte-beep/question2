```python
#=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=
#
#                                   ES335- Machine Learning- Assignment 1
#
# This file is used to create the dataset for the mini-project. The dataset is created by reading the data from
# the Combined folder. The data is then split into training, testing, and validation sets. This split is supposed
# to be used for all the modeling purposes.
#
#=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=

# Library imports
from sklearn.model_selection import train_test_split
import pandas as pd
import numpy as np
import os

# Constants
time = 10
offset = 100
folders = ["LAYING","SITTING","STANDING","WALKING","WALKING_DOWNSTAIRS","WALKING_UPSTAIRS"]
classes = {"WALKING":1,"WALKING_UPSTAIRS":2,"WALKING_DOWNSTAIRS":3,"SITTING":4,"STANDING":5,"LAYING":6}

combined_dir = os.path.join("Combined")

#=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-==-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=
                                                # Train Dataset
#=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-==-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=

X_train=[]
y_train=[]
dataset_dir = os.path.join(combined_dir,"Train")

for folder in folders:
    files = os.listdir(os.path.join(dataset_dir,folder))

    for file in files:

        df = pd.read_csv(os.path.join(dataset_dir,folder,file),sep=",",header=0)
        df = df[offset:offset+time*50]
        X_train.append(df.values)
        y_train.append(classes[folder])

X_train = np.array(X_train)
y_train = np.array(y_train)


#=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-==-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=
                                                # Test Dataset
#=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-==-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=

X_test=[]
y_test=[]
dataset_dir = os.path.join(combined_dir,"Test")

for folder in folders:
    files = os.listdir(os.path.join(dataset_dir,folder))
    for file in files:

        df = pd.read_csv(os.path.join(dataset_dir,folder,file),sep=",",header=0)
        df = df[offset:offset+time*50]
        X_test.append(df.values)
        y_test.append(classes[folder])

X_test = np.array(X_test)
y_test = np.array(y_test)

#=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-==-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=
                                                # Final Dataset
#=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-==-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=

# USE THE BELOW GIVEN DATA FOR TRAINING, TESTING, AND VALIDATION purposes

# concatenate the training and testing data
X = np.concatenate((X_train,X_test))
y = np.concatenate((y_train,y_test))

# split the data into training,testing, and validation sets
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.4,random_state=4,stratify=y)
X_test,X_val,y_test,y_val = train_test_split(X_test,y_test,test_size=0.5,random_state=4,stratify=y_test)

print("Training data shape: ",X_train.shape)
print("Testing data shape: ",X_test.shape)
print("Validation data shape: ",X_val.shape)

#=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-==-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=
```

```python
# in this part of the code we are creating the numpy array for squared sum of the linear acceleration for the training data.
acceleration_columns = X_train[:, :, :3]

# Compute the squared sum of acx^2 + acy^2 + acz^2 for each entry in X_train
linear_acceleration_train = np.sum(acceleration_columns ** 2, axis=2)

# Display the result
print(linear_acceleration_train)
```

```python
import numpy as np
from sklearn.utils import shuffle

# Assuming linear_acceleration_train and y_train are already defined

# Ensure y_train is a 1D array
y_train = y_train.flatten()

# Combine linear acceleration with activity labels
labeled_dataset = np.column_stack((linear_acceleration_train, y_train))
```

```python
# As we can see labeled_dataset contains the acceleration net along with the corresponding label
# for example all the values (except the last one which is the lable itself) in the first subarray has the label 5, Similarly 
# in the second subarray it also has a label 5 and so on...
# then we are suffling the dataset in order to prevent any kind of overfitting
print(labeled_dataset)
shuffled_dataset = shuffle(labeled_dataset, random_state=42)
print("\n")
print(shuffled_dataset)
```

```python
# now since we are only interested in the static vs dynamic state i.e binary classification, we are making a y_binary array 
# with static activities being labeled 0 and dynamic as 1.
X_final = shuffled_dataset[:, :-1]
y_final = shuffled_dataset[:, -1]
y_binary = np.where(np.isin(y_final, [1, 2, 3]), 0, 1)
print(X_final)
print("\n")
print(y_binary)
```

```python
# we can create the below given histogram and see how there is a clear distinction between the dynamic and the static values. 
# inference: if the linear acceleration has a value anything other than close to 1 than we can say that it is a staic acitivity.
import matplotlib.pyplot as plt

# Extract linear acceleration values for static and dynamic activities
linear_acceleration_static = X_final[y_binary == 0]
linear_acceleration_dynamic = X_final[y_binary == 1]

# Plot histograms
plt.hist(linear_acceleration_static.flatten(), bins=50, alpha=0.5, label='Static')
plt.hist(linear_acceleration_dynamic.flatten(), bins=50, alpha=0.5, label='Dynamic')
plt.xlabel('Linear Acceleration Value')
plt.ylabel('Frequency')
plt.legend()
plt.show()
```

```python
# the below given kernel density plot can also help us draw the same inferences
import seaborn as sns
import matplotlib.pyplot as plt

# Extract linear acceleration values for static and dynamic activities
linear_acceleration_static = X_final[y_binary == 0].flatten()
linear_acceleration_dynamic = X_final[y_binary == 1].flatten()

# Create kernel density plot
sns.kdeplot(linear_acceleration_static, label='Static', fill=True)
sns.kdeplot(linear_acceleration_dynamic, label='Dynamic', fill=True)

plt.xlabel('Linear Acceleration Value')
plt.ylabel('Density')
plt.legend()
plt.show()
```

```python
# we can also use box plots and violin plots for any further inferences.
import seaborn as sns
import matplotlib.pyplot as plt

# Create a DataFrame for easy plotting
data = {'Activity': np.repeat(['Static', 'Dynamic'], [len(linear_acceleration_static), len(linear_acceleration_dynamic)]),
        'Linear Acceleration': np.concatenate([linear_acceleration_static, linear_acceleration_dynamic])}

df = pd.DataFrame(data)

# Violin plot
plt.figure(figsize=(10, 6))
sns.violinplot(x='Activity', y='Linear Acceleration', data=df, inner='quartile', palette='pastel')
plt.title('Violin Plot - Linear Acceleration for Static and Dynamic Activities')
plt.show()

# Box plot
plt.figure(figsize=(10, 6))
sns.boxplot(x='Activity', y='Linear Acceleration', data=df, palette='pastel')
plt.title('Box Plot - Linear Acceleration for Static and Dynamic Activities')
plt.show()
```
